```python
# Divide the 'gross' and 'budget' columns by 1000000 to convert '$' to 'million $'
movies.budget=movies.budget.apply(lambda x:x/1000000)
movies.Gross=movies.Gross.apply(lambda x:x/1000000)
movies.head()
```

```python
# Create the new column named 'profit' by subtracting the 'budget' column from the 'gross' column
def profit(x):
    bud=x["budget"]
    gross=x["Gross"]
    return (gross-bud)
```

```python
movies["Profit"]=movies.apply(profit,axis=1)
movies.head()
```

```python
# Sort the dataframe with the 'profit' column as reference using the 'sort_values' function. Make sure to set the argument
#'ascending' to 'False'
top10=movies.sort_values(by="Profit",ascending=False)
top10.head(6)
```

```python
# Get the top 10 profitable movies by using position based indexing. Specify the rows till 10 (0-9)

top10.iloc[0:10]
```

```python
#Plot profit vs budget
sns.jointplot(top10.budget,top10.Profit)
plt.show()
```

```python
# Change the scale of MetaCritic
movies.MetaCritic=movies.MetaCritic.apply(lambda x:x/10)
movies.MetaCritic
```

```python
# Find the average ratings
def avg(x):
    meta=x["MetaCritic"]
    imdb=x["IMDb_rating"]
    average=(meta+imdb)/2
    return average
```

```python
movies["Avg_rating"]=movies.apply(avg,axis=1)
movies.head()
```

```python
def difference(x):
    meta=x["MetaCritic"]
    imdb=x["IMDb_rating"]
    diff=meta-imdb
    return abs(diff)
```

```python
movies["Absolute_Value"]=movies.apply(difference,axis=1)
movies.head()
```

```python
df_1=movies[movies.Absolute_Value<0.5]
```

```python
# Group the movies by individual genres
#df_by_genre.info()
df_by_g1=df_by_genre.groupby("genre_1").sum()
df_by_g1
```

```python
# Add the grouped data frames and store it in a new data frame
df_by_g2=df_by_genre.groupby("genre_2").sum()
df_by_g2

```

```python
df_by_g3=df_by_genre.groupby("genre_3").sum()
df_by_g3
```

```python
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```
